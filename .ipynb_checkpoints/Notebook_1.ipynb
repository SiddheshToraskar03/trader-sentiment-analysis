```python
# FIXED STATISTICAL SIGNIFICANCE TESTING
print("=== STATISTICAL SIGNIFICANCE TESTING ===")

actual_sentiment_order = sentiment_stats.index.tolist()

# ANOVA Test Difference in PnL across sentiments
sentiment_groups = [merged_df[merged_df['classification'] == s]['Closed PnL'].dropna()
                   for s in actual_sentiment_order]

f_stat, p_value = stats.f_oneway(*sentiment_groups)
print(f"1. ANOVA Test - PnL across Sentiments:")
print(f"   F-statistic: {f_stat:.4f}, p-value: {p_value:.4f}")

if p_value < 0.05:
    print("   → SIGNIFICANT: PnL differs across sentiment categories (p < 0.05)")
else:
    print("   → NOT SIGNIFICANT: No difference in PnL across sentiments")

# Correlation between sentiment value and trade metrics
correlation_pnl = merged_df['value'].corr(merged_df['Closed PnL'])
correlation_size = merged_df['value'].corr(merged_df['Size USD'])

print(f"\n2. Correlation Analysis:")
print(f"   Sentiment vs PnL: {correlation_pnl:.4f}")
print(f"   Sentiment vs Trade Size: {correlation_size:.4f}")

# Compare first and last sentiment categories if available
if len(actual_sentiment_order) >= 2:
    sentiment1_pnl = merged_df[merged_df['classification'] == actual_sentiment_order[0]]['Closed PnL']
    sentiment2_pnl = merged_df[merged_df['classification'] == actual_sentiment_order[-1]]['Closed PnL']

    if len(sentiment1_pnl) > 0 and len(sentiment2_pnl) > 0:
        t_stat, p_val = stats.ttest_ind(sentiment1_pnl, sentiment2_pnl, equal_var=False)
        print(f"\n3. T-test: {actual_sentiment_order[0]} vs {actual_sentiment_order[-1]}:")
        print(f"   t-statistic: {t_stat:.4f}, p-value: {p_val:.4f}")

        if p_val < 0.05:
            print("   → SIGNIFICANT: Performance differs between these sentiments")
        else:
            print("   → No significant difference between these sentiments")
```

```python
# COMPREHENSIVE VISUALIZATIONS
def to_datetime_auto(series, local_tz='Asia/Kolkata'):
    """Parse timestamps that might be epoch (s/ms/us/ns) or strings."""
    s = series.copy()
    as_num = pd.to_numeric(s, errors='coerce')

    if as_num.notna().any():
        mx = as_num.dropna().abs().max()
        if mx < 1e11:   unit = 's'
        elif mx < 1e14: unit = 'ms'
        elif mx < 1e17: unit = 'us'
        else:           unit = 'ns'

        dt = pd.to_datetime(as_num, unit=unit, errors='coerce', utc=True)
        try:
            dt = dt.dt.tz_convert(local_tz).dt.tz_localize(None)
        except Exception:
            dt = dt.dt.tz_localize(None)
        return dt

    return pd.to_datetime(s, errors='coerce')

def get_colors_for_sentiments(sentiments):
    cmap = {
        'Extreme Fear': 'darkred',
        'Fear': 'red',
        'Neutral': 'gray',
        'Greed': 'green',
        'Extreme Greed': 'darkgreen'
    }
    return [cmap.get(s, 'gray') for s in sentiments]

merged_df = merged_df.copy()
merged_df['Timestamp'] = to_datetime_auto(merged_df['Timestamp'])
merged_df = merged_df.dropna(subset=['Timestamp']).sort_values('Timestamp')

for col in ['Closed PnL', 'Size USD', 'Execution Price']:
    if col in merged_df.columns:
        merged_df[col] = pd.to_numeric(merged_df[col], errors='coerce')

if 'fg_clean' in globals() and not fg_clean.empty:
    fg_clean = fg_clean.copy()

    if 'date' in fg_clean.columns:
        fg_clean['date'] = pd.to_datetime(fg_clean['date'], errors='coerce')
        fg_clean = fg_clean.dropna(subset=['date']).sort_values('date')
    elif 'Date' in fg_clean.columns:
        fg_clean['date'] = pd.to_datetime(fg_clean['Date'], errors='coerce')
        fg_clean = fg_clean.dropna(subset=['date']).sort_values('date')

    if 'classification' in fg_clean.columns:
        fg_clean['classification'] = (
            fg_clean['classification'].astype(str).str.strip().str.title()
        )

plt.style.use('seaborn-v0_8')
fig = plt.figure(figsize=(20, 16))

# Daily Trading Activity Over Time
plt.subplot(3, 3, 1)
if merged_df.empty:
    plt.text(0.5, 0.5, 'No valid timestamps after parsing',
             ha='center', va='center', transform=plt.gca().transAxes)
else:
    daily_trades = merged_df.set_index('Timestamp').resample('D').size()
    plt.plot(daily_trades.index, daily_trades.values, alpha=0.9, linewidth=2)
plt.title('Daily Trading Activity Over Time', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Number of Trades')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# Distribution of Closed PnL
plt.subplot(3, 3, 2)
sns.histplot(merged_df['Closed PnL'].dropna(), kde=True, bins=50, color='skyblue', edgecolor='black')
if merged_df['Closed PnL'].notna().any():
    plt.axvline(merged_df['Closed PnL'].mean(), color='red', linestyle='--',
                label=f"Mean: ${merged_df['Closed PnL'].mean():.2f}")
    plt.axvline(merged_df['Closed PnL'].median(), color='green', linestyle='--',
                label=f"Median: ${merged_df['Closed PnL'].median():.2f}")
plt.title('Distribution of Closed PnL', fontsize=14, fontweight='bold')
plt.xlabel('Closed PnL ($)')
plt.ylabel('Frequency')
plt.legend()

# Trade Size Distribution
plt.subplot(3, 3, 3)
sns.boxplot(data=merged_df, y='Size USD', color='lightgreen')
plt.title('Trade Size Distribution', fontsize=14, fontweight='bold')
plt.ylabel('Trade Size ($)')

# Trade Side Distribution
plt.subplot(3, 3, 4)
if 'Side' in merged_df.columns:
    side_counts = merged_df['Side'].value_counts()
    if len(side_counts) > 0:
        plt.pie(side_counts.values, labels=side_counts.index, autopct='%1.1f%%',
                colors=['lightblue', 'lightcoral'], startangle=90)
    else:
        plt.text(0.5, 0.5, 'No side data', ha='center', va='center', transform=plt.gca().transAxes)
else:
    plt.text(0.5, 0.5, 'Side column missing', ha='center', va='center', transform=plt.gca().transAxes)
plt.title('Trade Side Distribution', fontsize=14, fontweight='bold')

# Fear & Greed Index Distribution
plt.subplot(3, 3, 5)
if 'value' in fg_clean.columns:
    sns.histplot(fg_clean['value'].dropna(), bins=30, kde=True, color='orange', edgecolor='black')
    if fg_clean['value'].notna().any():
        plt.axvline(fg_clean['value'].mean(), color='red', linestyle='--',
                    label=f"Mean: {fg_clean['value'].mean():.1f}")
else:
    plt.text(0.5, 0.5, 'F&G values missing', ha='center', va='center', transform=plt.gca().transAxes)
plt.title('Fear & Greed Index Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Fear & Greed Value')
plt.ylabel('Frequency')
plt.legend()

# Feature Correlation Heatmap
plt.subplot(3, 3, 6)
numeric_cols = [c for c in ['Closed PnL', 'Size USD', 'Execution Price', 'value'] if c in merged_df.columns]
if 'value' not in merged_df.columns and 'value' in fg_clean.columns:
    temp = merged_df.copy()
    temp['DateOnly'] = merged_df['Timestamp'].dt.date
    v = fg_clean[['date', 'value']].copy()
    v['DateOnly'] = v['date'].dt.date
    temp = temp.merge(v[['DateOnly', 'value']], on='DateOnly', how='left')
    numeric_cols = [c for c in ['Closed PnL', 'Size USD', 'Execution Price', 'value'] if c in temp.columns]
    corr_mat = temp[numeric_cols].corr()
else:
    corr_mat = merged_df[numeric_cols].corr()
sns.heatmap(corr_mat, annot=True, cmap='coolwarm', center=0, square=True, fmt='.2f', cbar_kws={'shrink': 0.8})
plt.title('Feature Correlation Heatmap', fontsize=14, fontweight='bold')

# Cumulative PnL Over Time
plt.subplot(3, 3, 7)
if merged_df.empty:
    plt.text(0.5, 0.5, 'No valid timestamps after parsing',
             ha='center', va='center', transform=plt.gca().transAxes)
else:
    cumulative_pnl = (
        merged_df.set_index('Timestamp')['Closed PnL']
                 .fillna(0)
                 .resample('D')
                 .sum()
                 .cumsum()
    )
    plt.plot(cumulative_pnl.index, cumulative_pnl.values, linewidth=2)
plt.title('Cumulative PnL Over Time', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Cumulative PnL ($)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# Trader Performance Segments
plt.subplot(3, 3, 8)
if 'trader_segments' in globals() and not trader_segments.empty and 'Performance_Segment' in trader_segments.columns:
    segment_counts = trader_segments['Performance_Segment'].value_counts()
    colors = ['green', 'lightgreen', 'yellow', 'orange', 'red']
    plt.bar(segment_counts.index, segment_counts.values, color=colors[:len(segment_counts)], edgecolor='black', alpha=0.8)
else:
    plt.text(0.5, 0.5, 'No trader segment data', ha='center', va='center', transform=plt.gca().transAxes)
plt.title('Trader Performance Segments', fontsize=14, fontweight='bold')
plt.xlabel('Performance Segment')
plt.ylabel('Number of Traders')
plt.xticks(rotation=45)

# Daily Sentiment Distribution (from F&G)
plt.subplot(3, 3, 9)
if 'classification' in fg_clean.columns and not fg_clean.empty:
    sentiment_changes = (
        fg_clean.set_index('date')['classification']
               .resample('D')
               .first()
               .dropna()
    )
    sentiment_counts = sentiment_changes.value_counts()
    plt.bar(sentiment_counts.index, sentiment_counts.values,
            color=get_colors_for_sentiments(sentiment_counts.index), alpha=0.8, edgecolor='black')
    plt.title('Daily Sentiment Distribution', fontsize=14, fontweight='bold')
    plt.xlabel('Market Sentiment')
    plt.ylabel('Number of Days')
    plt.xticks(rotation=45)
else:
    plt.text(0.5, 0.5, 'Insufficient data\nfor sentiment distribution',
             ha='center', va='center', transform=plt.gca().transAxes)
    plt.title('Daily Sentiment Distribution', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('/content/drive/MyDrive/ds_siddhesh_toraskar/outputs/comprehensive_eda.png',
            dpi=300, bbox_inches='tight')
plt.show()
```

```python
# ADVANCED STATISTICAL ANALYSIS & CONFUSION MATRIX
print("\n5. ADVANCED STATISTICAL ANALYSIS")
print("="*50)

from sklearn.metrics import confusion_matrix, classification_report
import scipy.stats as stats

# Performance Classification
def classify_performance(pnl):
    if pnl > 100:
        return 'High Profit'
    elif pnl > 0:
        return 'Low Profit'
    elif pnl > -50:
        return 'Small Loss'
    else:
        return 'Large Loss'

merged_df['Performance_Class'] = merged_df['Closed PnL'].apply(classify_performance)

# Create confusion matrix style analysis: Sentiment vs Performance
sentiment_performance_matrix = pd.crosstab(
    merged_df['classification'],
    merged_df['Performance_Class'],
    normalize='index'
) * 100

print("Sentiment vs Performance Matrix (%):")
print(sentiment_performance_matrix.round(1))

# Visualization of the matrix
plt.figure(figsize=(12, 8))
sns.heatmap(sentiment_performance_matrix, annot=True, fmt='.1f', cmap='YlOrRd',
            cbar_kws={'label': 'Percentage (%)'})
plt.title('Sentiment vs Performance Classification Matrix', fontsize=16, fontweight='bold')
plt.xlabel('Performance Class')
plt.ylabel('Market Sentiment')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/ds_siddhesh_toraskar/outputs/confusion_matrix_style.png',
            dpi=300, bbox_inches='tight')
plt.show()

# Statistical Tests
print("\n📊 ADVANCED STATISTICAL TESTS:")

_, p_normality = stats.normaltest(merged_df['Closed PnL'].dropna())
print(f"• Normality Test (PnL): p-value = {p_normality:.4f}")
print("  → PnL distribution is" + (" NOT normal" if p_normality < 0.05 else " normal"))

# Compare means between top and bottom sentiment performers
if len(sentiment_stats) >= 2:
    top_sentiment = sentiment_stats['Avg_PnL'].idxmax()
    bottom_sentiment = sentiment_stats['Avg_PnL'].idxmin()

    top_data = merged_df[merged_df['classification'] == top_sentiment]['Closed PnL']
    bottom_data = merged_df[merged_df['classification'] == bottom_sentiment]['Closed PnL']

    if len(top_data) > 0 and len(bottom_data) > 0:
        t_stat, t_pval = stats.ttest_ind(top_data, bottom_data, equal_var=False)
        print(f"• T-test ({top_sentiment} vs {bottom_sentiment}):")
        print(f"  t-statistic = {t_stat:.4f}, p-value = {t_pval:.4f}")
        print("  → Difference is" + (" significant" if t_pval < 0.05 else " not significant"))

# Risk Analysis
print(f"\n📈 RISK ANALYSIS:")
volatility_by_sentiment = merged_df.groupby('classification')['Closed PnL'].std()
for sentiment, vol in volatility_by_sentiment.items():
    sharpe = sentiment_stats.loc[sentiment, 'Avg_PnL'] / vol if vol > 0 else 0
    print(f"• {sentiment:15}: Volatility = ${vol:.2f}, Sharpe = {sharpe:.3f}")
```

```python
# TIME SERIES ANALYSIS
print("\n6. TIME SERIES ANALYSIS")
print("="*50)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def to_datetime_auto(series, local_tz='Asia/Kolkata'):
    s = series.copy()
    as_num = pd.to_numeric(s, errors='coerce')
    if as_num.notna().any():
        mx = as_num.dropna().abs().max()
        if mx < 1e11:   unit = 's'
        elif mx < 1e14: unit = 'ms'
        elif mx < 1e17: unit = 'us'
        else:           unit = 'ns'
        dt = pd.to_datetime(as_num, unit=unit, errors='coerce', utc=True)
        try:
            dt = dt.dt.tz_convert(local_tz).dt.tz_localize(None)
        except Exception:
            dt = dt.dt.tz_localize(None)
        return dt
    return pd.to_datetime(s, errors='coerce')

merged_df = merged_df.copy()
merged_df['Timestamp'] = to_datetime_auto(merged_df['Timestamp'])
merged_df = merged_df.dropna(subset=['Timestamp']).sort_values('Timestamp')

for c in ['Closed PnL', 'Size USD']:
    if c in merged_df.columns:
        merged_df[c] = pd.to_numeric(merged_df[c], errors='coerce').fillna(0)

trade_count_col = 'Trade_Id' if 'Trade_Id' in merged_df.columns else 'Closed PnL'

trades_daily = (
    merged_df.set_index('Timestamp').resample('D').agg(
        Daily_PnL_Total=('Closed PnL', 'sum'),
        Daily_PnL_Avg=('Closed PnL', 'mean'),
        Daily_Trades=(trade_count_col, 'count'),
        Daily_Volume=('Size USD', 'sum')
    )
)

sentiment_daily = None
if 'fg_clean' in globals() and not fg_clean.empty:
    fgc = fg_clean.copy()
    date_col = 'date' if 'date' in fgc.columns else ('Date' if 'Date' in fgc.columns else None)
    if date_col is not None and 'value' in fgc.columns:
        fgc['date'] = pd.to_datetime(fgc[date_col], errors='coerce')
        fgc = fgc.dropna(subset=['date']).sort_values('date')
        sentiment_daily = fgc.set_index('date')['value'].resample('D').mean().rename('Daily_Sentiment')

if sentiment_daily is not None:
    daily_metrics = trades_daily.join(sentiment_daily, how='left')

    daily_metrics['Daily_Sentiment'] = daily_metrics['Daily_Sentiment'].ffill()
else:
    daily_metrics = trades_daily
    daily_metrics['Daily_Sentiment'] = np.nan

if daily_metrics.empty:
    print("No daily metrics available after resampling. Check Timestamp parsing and input data.")
else:
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))

    ax1.plot(daily_metrics.index, daily_metrics['Daily_PnL_Total'], color='blue', alpha=0.7, label='Daily PnL')
    ax1.set_ylabel('Daily PnL ($)', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    ax1.legend(loc='upper left')

    if daily_metrics['Daily_Sentiment'].notna().any():
        ax1b = ax1.twinx()
        ax1b.plot(daily_metrics.index, daily_metrics['Daily_Sentiment'], color='red', alpha=0.8, linewidth=2, label='Sentiment')
        ax1b.set_ylabel('Fear & Greed Index', color='red')
        ax1b.tick_params(axis='y', labelcolor='red')
        ax1b.legend(loc='upper right')
    ax1.set_title('Daily PnL vs Market Sentiment', fontsize=14, fontweight='bold')

    ax2.bar(daily_metrics.index, daily_metrics['Daily_Volume'], color='green', alpha=0.7, label='Daily Volume')
    ax2.set_ylabel('Daily Volume ($)')
    ax2.set_title('Daily Trading Volume', fontsize=14, fontweight='bold')
    ax2.legend()

    ax3.plot(daily_metrics.index, daily_metrics['Daily_Trades'], color='purple', alpha=0.8, linewidth=2, label='Daily Trades')
    ax3.set_ylabel('Number of Trades')
    ax3.set_xlabel('Date')
    ax3.set_title('Daily Trading Activity', fontsize=14, fontweight='bold')
    ax3.legend()

    plt.tight_layout()
    plt.savefig('/content/drive/MyDrive/ds_siddhesh_toraskar/outputs/time_series_analysis.png',
                dpi=300, bbox_inches='tight')
    plt.show()

if daily_metrics['Daily_Sentiment'].notna().any():
    sentiment_corr = daily_metrics[['Daily_Sentiment', 'Daily_PnL_Total', 'Daily_Volume', 'Daily_Trades']].corr()['Daily_Sentiment']
    print("\n📊 SENTIMENT CORRELATIONS:")
    print(f"• Sentiment vs Daily PnL:    {sentiment_corr.get('Daily_PnL_Total', np.nan):.3f}")
    print(f"• Sentiment vs Daily Volume: {sentiment_corr.get('Daily_Volume', np.nan):.3f}")
    print(f"• Sentiment vs Daily Trades: {sentiment_corr.get('Daily_Trades', np.nan):.3f}")
else:
    print("\nNo sentiment values aligned with trading days; correlations skipped.")
```

```python
# MACHINE LEARNING READINESS ANALYSIS
print("\n7. MACHINE LEARNING READINESS")
print("="*50)

from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

# Data for feature importance
ml_data = merged_df[['Size USD', 'value', 'Side', 'classification', 'Closed PnL']].copy()
ml_data = ml_data.dropna()

# Encode categorical
le_side = LabelEncoder()
le_sentiment = LabelEncoder()
ml_data['Side_encoded'] = le_side.fit_transform(ml_data['Side'])
ml_data['Sentiment_encoded'] = le_sentiment.fit_transform(ml_data['classification'])

# Feature importance using Random Forest
X = ml_data[['Size USD', 'value', 'Side_encoded', 'Sentiment_encoded']]
y = ml_data['Closed PnL']

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X, y)

feature_importance = pd.DataFrame({
    'feature': ['Trade Size', 'Sentiment Value', 'Trade Side', 'Sentiment Category'],
    'importance': rf.feature_importances_
}).sort_values('importance', ascending=False)

print("FEATURE IMPORTANCE FOR PnL PREDICTION:")
for _, row in feature_importance.iterrows():
    print(f"• {row['feature']:20}: {row['importance']:.3f}")

# Visualization of feature importance
plt.figure(figsize=(10, 6))
sns.barplot(data=feature_importance, x='importance', y='feature', palette='viridis')
plt.title('Feature Importance for PnL Prediction', fontsize=14, fontweight='bold')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/ds_siddhesh_toraskar/outputs/feature_importance.png',
            dpi=300, bbox_inches='tight')
plt.show()
```
